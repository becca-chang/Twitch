```python
import pandas as pd
import time
import urllib.parse
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
```

## twitchmetrics

```python
driver = webdriver.Chrome()
```

```python
url = "https://www.twitchmetrics.net/channels/viewership?lang=en"
driver.get(url)
streamers = driver.find_elements(By.CSS_SELECTOR, ".list-group-item h5.mb-0")
language_cats = driver.find_elements(By.CLASS_NAME, "mr-3")
streamer_names = []
languages = []
categories = []
for element in streamers:
    streamer_names.append(element.text)
for index, language_cat in enumerate(language_cats):
    if index % 2 == 0:
        languages.append(language_cat.text)
    else:
        categories.append(language_cat.text)
```

```python
df = pd.DataFrame(data={"Name": streamer_names, "Language": languages, "Category": categories})
df.to_csv("twitchmetrics_channels.csv")
```

```python
url = "https://www.twitchmetrics.net/games/viewership"
driver.get(url)
categories = driver.find_elements(By.CSS_SELECTOR, ".list-group-item h5")
category_list = []
for cat in categories:
    category_list.append(cat.text)
df = pd.DataFrame(data={"Category": category_list})
df.to_csv("twitchmetrics_categories.csv")
```

```python
import re
def remove_punctuation_from_directory(name):
    # Use a regular expression to remove all punctuation
    return re.sub(r'[^\w\s-]', ' ', name)
```

```python
category_list = ["Just Chatting", "Fortnite"]
for cat in category_list:
    url_path = f"https://www.twitchmetrics.net/channels/viewership?game={urllib.parse.quote(cat)}&lang=en"
    print(url_path)
    driver.get(url_path)
    time.sleep(2)
    streamers = driver.find_elements(By.CSS_SELECTOR, ".list-group-item h5.mb-0")
    streamer_names = []
    for element in streamers:
        streamer_names.append(element.text)
    file_name = remove_punctuation_from_directory(cat)
    df = pd.DataFrame(data={"Name": streamer_names})
    df.to_csv(f"1124_{file_name}.csv")
```

```python
driver.quit()
```
