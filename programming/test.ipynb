## twitchmetrics

```python
driver = webdriver.Chrome()
```

## Get Most Watched English Twitch Streamers

```python
url = "https://www.twitchmetrics.net/channels/viewership?lang=en"
driver.get(url)
streamers = driver.find_elements(By.CSS_SELECTOR, ".list-group-item h5.mb-0")
language_cats = driver.find_elements(By.CLASS_NAME, "mr-3")
streamer_names = []
languages = []
categories = []
for element in streamers:
    streamer_names.append(element.text)
for index, language_cat in enumerate(language_cats):
    if index % 2 == 0:
        languages.append(language_cat.text)
    else:
        categories.append(language_cat.text)
```

```python
df = pd.DataFrame(data={"Name": streamer_names, "Language": languages, "Category": categories})
df.to_csv("twitchmetrics_channels.csv")
```

## Get The Most Watched Games on Twitch,

```python
url = "https://www.twitchmetrics.net/games/viewership"
driver.get(url)
categories = driver.find_elements(By.CSS_SELECTOR, ".list-group-item h5")
category_list = []
for cat in categories:
    category_list.append(cat.text)
df = pd.DataFrame(data={"Category": category_list})
df.to_csv("twitchmetrics_categories.csv")
```

```python
df = pd.read_csv("test/641972806.csv", index_col=0)
```

```python
def re_message(df, column, **kwargs):
    df_size = len(df)
    df["subscribed_type"] = None
    df["cheer"] = None
    df["tier_level"] = None
    df["subscribed_month"] = None
    df["gifting_count"] = None
    df["re_message_error"] = None
    cheer_pattern = kwargs.get("cheer_pattern")
    subscribed_pattern = kwargs.get("subscribed_pattern")
    gifting_pattern = kwargs.get("gifting_pattern")
    messages = list(df[column].astype(str))
    for index, message in enumerate(messages):
        try:
            if re.match(cheer_pattern, message): # å°å¥‡é»
                df.loc[index, "subscribed_type"] = 3
                df.loc[index, "cheer"] = re.match(cheer_pattern, message).group(1)
            elif re.match(subscribed_pattern, message): # è‡ªå·±è¨‚é–±
                df.loc[index, "subscribed_type"] = 1
                df.loc[index, "tier_level"] = re.match(cheer_pattern, message).group(1)
                df.loc[index, "subscribed_month"] = re.match(cheer_pattern, message).group(2)
            elif re.match(gifting_pattern, message): # è´ˆé€è¨‚é–±
                df.loc[index, "subscribed_type"] = 2
                df.loc[index, "tier_level"] = re.match(cheer_pattern, message).group(1)
                df.loc[index, "gifting_count"] = re.match(cheer_pattern, message).group(2)
            else:
                df.loc[index, "subscribed_type"] = 0
        except Exception as e:
            df.loc[index, "re_message_error"] = e
            continue
```

```python
cheer_pattern = r"Cheer(\d+)(?:\s|$)"
subscribed_pattern = r"subscribed at Tier (\d+).*?(\d+|\w+) month"
gifting_pattern = r"gifting (\d+) Tier (\d+) Subs to (\w+)'s community"
re_message(df, 
           "message",
           **{"cheer_pattern": cheer_pattern,
              "subscribed_pattern": subscribed_pattern,
              "gifting_pattern": gifting_pattern
            }
)
```
